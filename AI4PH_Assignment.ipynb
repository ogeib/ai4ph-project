```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("data/deaths_fatality_type.csv")


print("First 5 rows of the dataset:")
display(df.head())


print("\nDataset Information:")
df.info()


print("\nSummary Statistics:")
display(df.describe())


print("\nMissing Values in Each Column:")
print(df.isnull().sum())


plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Fatality Type', order=df['Fatality Type'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Distribution of Fatality Types")
plt.xlabel("Fatality Type")
plt.ylabel("Count")
plt.show()


print("\nSummary Statistics:")
print("Total Number of Entries:", df.shape[0])  # Total number of rows
print("Number of Unique Fatality Types:", df['Fatality Type'].nunique())  # Unique
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


df['date'] = pd.to_datetime(df['date'])


df = df.sort_values(by='date')


plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='date', y='deaths_total', marker='o')

plt.xticks(rotation=45)
plt.title("Total Deaths Over Time")
plt.xlabel("Date")
plt.ylabel("Total Deaths")
plt.grid(True)
plt.show()
```

## Summary of Findings  
- The dataset contains information on total deaths, COVID-related deaths, and other unknown causes.  

1. Trends in Total Deaths Over Time

The scatter plot visualization of total deaths over time indicates clear seasonal peaks and fluctuations. The following trends can be observed:

There are three prominent spikes in total deaths around 2021, 2022, and 2023.
The spikes correspond to known COVID-19 waves, suggesting an increase in fatalities during peak transmission periods.
After mid-2023, there is a steady decline in deaths, possibly due to improved public health interventions, vaccination efforts, and better treatment protocols.
While some fluctuations still exist in 2024, they are less pronounced than in earlier years.

2. Summary Statistics

The table summarizes key descriptive statistics for the dataset:

The mean (average) total deaths is 11.22, with a standard deviation of 13.83, indicating significant variation in daily reported deaths.
The minimum recorded total deaths is 0, while the maximum is 87, suggesting that some days had no reported fatalities, while others saw high peaks.
The median (50th percentile) total deaths is 6, meaning that half of the recorded values are below this number.
COVID-19-specific deaths (direct and contributing causes)
Mean COVID-19 deaths: 7.84, with a max value of 68.
Mean COVID-19 contributory deaths: 2.75, max of 22.
The presence of a “death_unknown_missing” category suggests potential gaps in reporting or classification errors.
